What US city had the highest number of Sales?

```python
#analysis['City'] = analysis['Purchase Address'] .apply(lambda x: x.split(',')[1] ) 1st method
#analysis

def get_city(address): #2nd method
    return address.split(',')[1]

def get_address(address):
    return address.split(',')[2].split(' ')[1]


analysis['City'] = analysis['Purchase Address'] .apply(lambda x: get_city(x)+ '(' +get_address(x)+')')
```

```python
analysis
```

```python
results = analysis.groupby('City').sum()
```

```python
results
```

```python
plt.figure(figsize=(10, 10), dpi=80)
cities = [city for city, df in analysis.groupby('City')]
plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation = 'vertical',size = 8)
plt.ylabel('Sales in Us Dollars')
plt.xlabel('City');
```

What products are most often sold together??

```python
analysis
```

```python
df = analysis[analysis['Order ID'].duplicated(keep=False)] # shows the data with same order ID
df['grouped'] = df.groupby('Order ID') ['Product'].transform(lambda x: ','.join(x))
df.head()
```

 Since we have the same order twice we need to drop one of them

```python
df = df[['Order ID', 'grouped']].drop_duplicates()
```

```python
df.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
for row in df ['grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
count.most_common(10)

for key , value in count.most_common(10):
    print(key, value)
```
